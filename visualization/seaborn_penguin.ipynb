```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
peng=sns.load_dataset('penguins')
```

```python
print(peng)
```

```python
sns.relplot(peng,x='species',y='body_mass_g',hue='sex')
```

```python
sns.relplot(peng,x='flipper_length_mm',y='body_mass_g',hue='species',height=12,aspect=0.5,col='island')
```

```python
sns.relplot(peng,x='flipper_length_mm',y='body_mass_g',hue='sex',height=10,col='island',row='species')
```

```python
sns.jointplot(peng,x='species',y='island')
```

```python
plt.xlabel('species')
plt.ylabel('island')
plt.scatter(peng['species'],peng['island'])
```
