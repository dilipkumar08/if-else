```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset('tips')
print(tips)
```

```python
sns.displot(data=tips,x='total_bill',kde=True)
```

```python
sns.lineplot(data=tips,x='total_bill',y='tip')
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker',row='day')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker',row='day',col='time')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker',row='day',col='time',palette=['grey','pink'])#,size='size
```

```python
sns.boxplot(tips['tip'])
sns.swarmplot(tips['tip'])
sns.violinplot(tips['tip'],color='pink')
```

```python
sns.pairplot(tips)
```
