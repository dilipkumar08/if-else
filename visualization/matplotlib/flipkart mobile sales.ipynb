```python
import matplotlib.pyplot as plt
import math
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('D:/python/dataset/Flipkart Mobile.csv')
```

```python
brand=[]
sales=[]
for i in set(data['brand']):
    brand.append(i)
    sales.append(sum(data.loc[data['brand']==i]['sales']))
plt.title('Flipkart Mobile sales data')
plt.xlabel('Brand')
plt.ylabel('Sales')
plt.grid()
plt.bar(brand,sales,color='green')
plt.show()
```

```python
plt.pie(sales,labels=brand,autopct='%1.0f%%')
plt.show()
```

```python
model=[]
model_sales=[]
for j in set(data.loc[data['brand']=='Realme']['model']):
    model.append(j)
    model_sales.append(sum(data.loc[data['model']==j]['sales']))
plt.figure(figsize=(26, 12)) 
plt.title('Flipkart Realme sales')
plt.xlabel('Model')
plt.ylabel('Sales')
plt.grid()
plt.bar(model,model_sales,color='c')

plt.show()
```

```python
plt.figure(figsize=(30, 15)) 
plt.pie(model_sales,labels=model,autopct='%1.0f%%')
plt.show()
```
